```python
import pandas as pd
import json
from pathlib import Path
```

```python
BASE_PATH = Path("/data/commonvoice/cv-corpus-9.0-2022-04-27")

LANG_CODE = "es"
```

```python
path  = BASE_PATH / LANG_CODE
```

```python
path
```

```python
# [x for x in (path / "clips").glob("**/*")]
```

```python
train = pd.read_csv(path/"train.tsv", sep="\t")
validated= pd.read_csv(path/"validated.tsv", sep="\t", low_memory=False)
dev = pd.read_csv(path/"dev.tsv", sep="\t", low_memory=False)
test = pd.read_csv(path/"test.tsv", sep="\t", low_memory=False)
```

```python
len(train), len(validated), len(dev), len(test)
```

```python
ids = validated["sentence"]
validated[ids.isin(ids[ids.duplicated()])].sort_values("sentence")
```

```python
validated[ids.isin(ids[ids.duplicated()])].sort_values("sentence").sentence.nunique()
```

```python
df = pd.DataFrame.merge(validated, test, on="path", how="outer", indicator=True)
full_train=df[df['_merge']=='left_only']
```

```python
full_train
```

```python
df = pd.DataFrame.merge(validated, dev, on="path", how="outer", indicator=True)

full_train=df[df['_merge']=='left_only']

full_train
```

```python
validated
```

```python
target = dev
pairs=[]

for _, row in target.iterrows():
    pair = {
        "file": row.path,
        "text": row.sentence,
    }

    pairs.append(pair)
```

```python
data = pairs

result = {
    "root": "/data/commonvoice/",
    "data": data
}
```

```python
result
```

```python
split = "dev"
output = f"out/{LANG_CODE}-{split}.json"
with open(f"{output}", "w") as f:
    json.dump(result, f)
```
